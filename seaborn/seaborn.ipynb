```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data=sns.load_dataset('tips') ## this is a data set of tips
data
```

```python
data.head()
```

```python
data['total_bill']
```

```python
sns.lineplot(data['total_bill'])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data) #
```

```python
sns.histplot(data['total_bill'])
```

```python
plt.hist(data['total_bill']) #  this is a histogram plot by using matplotlib
```

```python
sns.kdeplot(data['total_bill'])
```

```python
#catplot
#relplot
```

```python
sns.relplot(x="total_bill",y="tip",data=data)
```

```python
data['time'].unique()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style="smoker",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style="smoker",kind="scatter",row="time",data=data)
```

```python
sns.relplot(y="total_bill",x="tip",hue="time",style='smoker',kind="scatter",col="smoker",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style="smoker",kind="scatter",row="time",data=data)
```

```python
data["day"].unique()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',kind="scatter",col="day",data=data)
```

```python
sns.relplot(y="total_bill",x="tip",data=data)
plt.title("this is a scatter plot")
plt.xlabel("total_bill",fontsize=20,color="black")
plt.ylabel("tip",fontsize=20,color="green")
plt.show()
```

```python
sns.catplot(x="day",y='total_bill',data=data)
```

```python
sns.catplot(x="day",y='total_bill',kind="swarm",data=data)
```

```python
sns.catplot(x="day",y='total_bill',kind="strip",data=data)
```

```python
sns.catplot(x="day",y='total_bill',kind="box",data=data)
```

```python
sns.catplot(x="day",y='total_bill',kind="bar",data=data)
```

```python
sns.catplot(x="day",y='total_bill',kind="boxen",data=data)
```

```python
sns.catplot(x="day",y='total_bill',kind="violin",data=data)
```
